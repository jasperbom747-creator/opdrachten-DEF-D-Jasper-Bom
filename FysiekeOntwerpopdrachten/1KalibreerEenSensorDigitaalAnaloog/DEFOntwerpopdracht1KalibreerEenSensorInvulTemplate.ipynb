## *Opdracht 6*: Meten en fitten, Leerdoel

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

lux = np.array([7,12,15,18,23,29,33,42,52,65,84,105])
arduino = np.array([364,495,583,628,643,735,753,796,826,888,950,978])
```

```python
def model(x, a, b, c):
    return a * x**b + c

# Beginwaarden (belangrijk voor goede fit!)
initial_guess = [10, 0.5, 300]

popt, pcov = curve_fit(model, lux, arduino, p0=initial_guess, maxfev=10000)

a, b, c = popt


lux_fit = np.linspace(min(lux), max(lux), 100)
arduino_fit = model(lux_fit, a, b, c)


plt.figure()
plt.plot(lux,arduino, 'r.', label='Meetdata')
plt.plot(lux_fit, arduino_fit, '--', label='Power fit')
plt.xlabel('Lichtintensiteit in lux')
plt.ylabel('Sensorwaarde')
plt.title('De lichtintensiteit uitgezet tegen de sensorwaarde')
plt.legend()
plt.grid()
plt.show()


```

Voor het gebruik van deze sensor in een automatische straatlantaarn volstaan de metingen, de sensor kan een duidelijk onderschijd maken tussen hoge en lage lux waarden en zo met voldoende nauwkeurigheid bepalen of de lamp aan of uit moet staan.
